```python
# Open text file
f = open('federalist_papers.txt')

# Read the data into a text variable
data = f.read()
print(data)
```

```python
data = data.replace('\n', ' ')
data = data.lower()
words = data.split(' ')
print(words[:10])
```

```python
print(len(words))
```

```python
while_count = 0
whilst_count = 0
bob_count = 0
for word in words:
    if word.find('while') != -1:
        while_count = while_count + 1
    if word.find('whilst') != -1:
        whilst_count = whilst_count + 1
    if word.find('bob') != -1:    
        bob_count = bob_count + 1

print(while_count)
print(whilst_count)
print(bob_count)
```

```python
def strip_non_alpha_characters(word):
    # ASCII values for characters between lowercase 'a' and lowercase 'z' are between 97 and 122
    # Python function that converts characters to ascii is called ord()
    
    word = word.strip()
    
    clean_word = ''

    for letter in word:
        # Checking if the ascii code of the letter is between 97 and 122 (between 'a' and 'z')
        if ord(letter) >= 97 and ord(letter) <= 122:
            # if lower case letter, concat. to variable clean_word
            clean_word = clean_word + letter
    
    return clean_word
    
```

```python
freq_dictionary = {}

for word in words:
    word = strip_non_alpha_characters(word)
    if word in freq_dictionary.keys():
        freq_dictionary[word] = freq_dictionary[word] + 1
    else:
        freq_dictionary[word] = 1
```

```python
f = open('diabetes.csv')
#data = f.readlines()
#print(data)

data = f.read()
data_list = data.split('\n')
print(data_list)
```

```python
for row in data_list:
    print("|" + row + "|")
```

```python
columns = {}

# Create a dictionary where column name is the key and the columns index (position in the list) is the value
column_headers = data_list[0].split(',')
for i in range(0, len(column_headers)):
    column_name = column_headers[i]
    # Column name is the index,
    # Position in the list is the value
    columns[column_name] = i

print(columns)
```

```python
# We are essentially creating a list of lists.  The master list represents the entire 'table'.
# Each child list represents a row of data
matrix = []
for row in data_list[1:]:
    row_list = row.split(',')
    matrix.append(row_list)

print(matrix)
```

```python
# This will give me row 5 in the data
print(matrix[4])
```

```python
# Print a value in row 5 in 'gender' column

print(matrix[4][columns['gender']])
```

```python
# Create a function that returns a single column from a dataset
def get_column_from_data(dataset, column_name):
    column_data = []
    for row in dataset:
        if len(row) == len(columns):
            column_data.append(row[columns[column_name]])
    
    return column_data
```

```python
gender = get_column_from_data(matrix, 'gender')
print(gender)
```

```python
import statistics
```

```python
print(statistics.mode(gender))
```

```python
def get_frequencies_from_column(col_values):
    value_frequencies = {}
    for val in col_values:
        if val in value_frequencies.keys():
            value_frequencies[val] = value_frequencies[val] + 1
        else:
            value_frequencies[val] = 1
    
    return value_frequencies
```

```python
print(get_frequencies_from_column(gender))
```

```python
# Get the list of values for 'chol' column
chol = get_column_from_data(matrix, 'chol')

# Get frequency counts for values in 'chol' column
print(get_frequencies_from_column(chol))
```

```python
# Get the list of values for 'frame' column
frame = get_column_from_data(matrix, 'frame')

# Get frequency counts for values in 'frame' column
print(get_frequencies_from_column(frame))
```
