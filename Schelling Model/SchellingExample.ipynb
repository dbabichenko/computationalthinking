```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
import random
```

```python
w, h = 10, 10
cells = ['X','O','E']
MAX_EMPTY_CELL_PERCENTAGE = 0.2
empty_cells = 0
max_empty_cells = (w * h) * MAX_EMPTY_CELL_PERCENTAGE
# Create a 50-dimensional grid
grid = [['E' for x in range(w)] for y in range(h)]

# Place agents on the grid, leaving 10% of cells empty
for i in range(w):
    for k in range(h):
        cell_value = random.choice(cells)
        grid[i][k] = cell_value
        empty_cells += 1
        if empty_cells >= max_empty_cells:
            cells = ['X','O']
            
```
