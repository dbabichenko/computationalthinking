```python
import pandas as pd
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def get_node_colors(graph):
    color_map = []
    for node in graph:
        if 'G' in node:
            color_map.append('red')
        elif 'P' in node: 
            color_map.append('green')   
        elif 'H' in node:
            color_map.append('yellow')   
    return color_map
```

```python
# Creating a Graph 
G = nx.Graph() # Right now G is empty
```

```python
# Add a node
G.add_node('G0') 
```

```python
# Add multiple nodes from a list
G.add_nodes_from(['G1','G2','G3'])
```

```python
G.add_edge('G0','G1')
G.add_edge('G0','G2')
G.add_edge('G0','G3')
```

```python
nx.draw(G, with_labels=True, node_color=get_node_colors(G), node_size=700)
```

```python
G.add_nodes_from(['P1','P2','P3'])
G.add_nodes_from(['H1','H2'])
G.add_edge('G0','P1')
G.add_edge('G0','P2')
G.add_edge('G0','P3')
G.add_edge('G0','H1')
G.add_edge('G0','H2')
nx.draw(G, with_labels=True, node_color=get_node_colors(G), node_size=700)
```
